## Trust Propagation - deviation :attack profile detection

```python
import pandas as pd
```

```python
trust = pd.read_csv('trust.txt', sep=" ", header=None)
trust.columns = ['trustor', 'trusting','t_value']
ratings = pd.read_csv('ratings.txt', sep=" ", header=None)
ratings.columns = ['user', 'item','rating']
ratings.iloc[100,:]
```

```python
print (ratings.shape)
print (trust.shape)
```

```python
# ratings['user'].value_counts()
```

```python
# auth_ratings = ratings[ratings['user'].value_counts()>150]
filtered = ratings.groupby('user')['user'].filter(lambda x: len(x) >= 150)
```

```python
print (filtered.shape)
filtered.value_counts()
```

```python
# type(filtered)
# ratings[ratings['user'].isin(filtered)]
```

```python
auth_user = set(filtered.tolist())
print (auth_user)       #set of auth users
```

```python
for u in auth_user:
#     print (trust[trust['trustor']==u].shape[0])
    trustee = trust[trust['trustor']==u]['trusting'].tolist()   #set of trustee of user
    l = ratings[ratings['user']==u]['item'].tolist()   #items rated by user U
    for item in l:
        fil_i = list()
        fil_i.append(ratings[ratings['item']==item]['user'].tolist())  #users rated the item which is rated by user U
        p = 0
        for i in fil_i:
            if(i in trustee):
                p = p+ trust[trust['trusting']==i]['t_val']*ratings[(ratings['user']==i) & (ratings['item']==item)]
        print (ratings[(ratings['user']==u) & (ratings['item']==item)],'    ',p)
```
